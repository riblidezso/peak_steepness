```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
OldPredDF = pd.read_csv('preds/original_model_preds.csv')
NewPredDF = pd.read_csv('preds/slightly_modified_model_preds.csv')
```

```python
print(((OldPredDF.pred_omega - OldPredDF.true_omega)**2).mean()**(1./2))
print(((OldPredDF.pred_sigma - OldPredDF.true_sigma)**2).mean()**(1./2))
```

```python
print(((NewPredDF.pred_omega - NewPredDF.true_omega)**2).mean()**(1./2))
print(((NewPredDF.pred_sigma - NewPredDF.true_sigma)**2).mean()**(1./2))
```

```python
plt.rcParams['figure.figsize'] = (14, 13)
labelFS = 25
#plt.suptitle('Improved model', fontsize=25)
plt.subplot('221')
plt.scatter(x=OldPredDF.true_omega, y=OldPredDF.pred_omega, s=10, c='r')
plt.plot([0.1, 0.8], [0.1, 0.8], 'k-')
plt.vlines(0.26, 0, 0.6, linestyles='--', label='fiducial')
plt.text(x=0.17, y=0.65, fontdict={'fontsize':40}, s='Gupta et al.')
plt.xlabel(r'True $\Omega_m$', fontsize=labelFS)
plt.ylabel(r'Predicted $\Omega_m$', fontsize=labelFS)
plt.tick_params(axis='both', which='major', labelsize=17)
plt.xlim(0.15, 0.75)
plt.ylim(0.15, 0.75)

plt.subplot('222')
plt.scatter(x=OldPredDF.true_sigma, y=OldPredDF.pred_sigma, s=10, c='r')
plt.plot([0, 1.4], [0, 1.4], 'k-')
#plt.title(r'$\sigma_8$', fontsize=20)

plt.xlabel(r'True $\sigma_8$', fontsize=labelFS)
plt.ylabel(r'Predicted $\sigma_8$', fontsize=labelFS)
plt.vlines(0.8, 0, 2, linestyles='--', label='fiducial')
plt.tick_params(axis='both', which='major', labelsize=17)
#plt.text(x=0.07, y=1.1, fontdict={'fontsize':40}, s='B')

plt.xlim(0, 1.3)
plt.ylim(0, 1.3)

plt.subplot('223')
plt.scatter(x=NewPredDF.true_omega, y=NewPredDF.pred_omega, s=10, c='b')
plt.plot([0.1, 0.8], [0.1, 0.8], 'k-')
plt.vlines(0.26, 0, 0.6, linestyles='--', label='fiducial')
#plt.title(r'$\Omega_m$', fontsize=20)
plt.xlabel(r'True $\Omega_m$', fontsize=labelFS)
plt.ylabel(r'Predicted $\Omega_m$', fontsize=labelFS)
plt.tick_params(axis='both', which='major', labelsize=17)
plt.text(x=0.17, y=0.65, fontdict={'fontsize':40}, s='This work')


plt.xlim(0.15, 0.75)
plt.ylim(0.15, 0.75)

plt.subplot('224')
plt.scatter(x=NewPredDF.true_sigma, y=NewPredDF.pred_sigma, s=10, c='b')
plt.plot([0, 1.4], [0, 1.4], 'k-')
#plt.title(r'$\sigma_8$', fontsize=20)

plt.xlabel(r'True $\sigma_8$', fontsize=labelFS)
plt.ylabel(r'Predicted $\sigma_8$', fontsize=labelFS)
plt.vlines(0.8, 0, 2, linestyles='--', label='fiducial')
plt.tick_params(axis='both', which='major', labelsize=17)
#plt.text(x=0.07, y=1.1, fontdict={'fontsize':40}, s='D')

plt.xlim(0, 1.3)
plt.ylim(0, 1.3)

plt.savefig('figs/final/model_comp_scatter.png', transparent=True, bbox_inches='tight', pad_inches=0)
plt.show()
```
