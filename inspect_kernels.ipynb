# Inspect the low level kernels to learn about the representation used by a neural network

---

We trained a model similar to the main one reported in the paper, with 7x7 bottom kernels and  L2 regularization in order to achieve kernels which are easier to interpret. (This particular model also includes batch normalization.)

The model was trained on lower resolution data (2 arcmin), in order to achieve relevant results for measurements.

For more details see the paper.

---

```python
%pylab inline
```

### Load a model with 7x7 bottom conv

Ignore warning

```python
from keras.models import load_model
net = load_model('data/7x7_2arcmin.model')
```

#### Get the weights

```python
w = net.get_weights()
```

### Plot first convolutional layer weights

- Most kernels are rather featureless (due to L2 regularization)

```python
for i in xrange(32):
    imshow(w[0][:,:,0,i],vmin=-0.05,vmax=0.05,cmap='BrBG')
    colorbar()
    title(str(i))
    show()
```

# Two kernels show particularly strong patterns 

### [Discrete Laplace filter](https://en.wikipedia.org/wiki/Discrete_Laplace_operator)



```python
imshow(w[0][:,:,0,30],vmin=-0.05,vmax=0.05,cmap='BrBG')
colorbar()
```

### [A Roberts cross kernel](https://en.wikipedia.org/wiki/Roberts_cross)




```python
imshow(w[0][:,:,0,25],vmin=-0.05,vmax=0.05,cmap='BrBG')
colorbar()
```

---

# Conclusions

Most of the kernels are rather featureless due to strong L2 regularization.

Two strong and distinct patterns stand out, the discrete Laplace filter abd a Roberts cross kernel. 

Both characteristic kernels describe the gradient of the field when applied around a peak. We might have found something.
